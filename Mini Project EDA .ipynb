```python
#Combining maritial status
dupl_df['Marital_Status'] = dupl_df['Marital_Status'].replace(['Married', 'Together'],'relationship')
dupl_df['Marital_Status'] = dupl_df['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'],'Single')
```

```python
dupl_df['Marital_Status'].value_counts()  
```

```python
# Combining columns
dupl_df['Kids'] = dupl_df['Kidhome'] + dupl_df['Teenhome']
dupl_df['Expenses'] = dupl_df['MntWines'] + dupl_df['MntFruits'] + dupl_df['MntMeatProducts'] + dupl_df['MntFishProducts'] + dupl_df['MntSweetProducts'] + dupl_df['MntGoldProds']
dupl_df['TotalAcceptedCmp'] = dupl_df['AcceptedCmp1'] + dupl_df['AcceptedCmp2'] + dupl_df['AcceptedCmp3'] + dupl_df['AcceptedCmp4'] + dupl_df['AcceptedCmp5'] + dupl_df['Response']
dupl_df['NumTotalPurchases'] = dupl_df['NumWebPurchases'] + dupl_df['NumCatalogPurchases'] + dupl_df['NumStorePurchases'] + dupl_df['NumDealsPurchases']
```

```python
#Dropping remaining columns
dupl_df=dupl_df.drop(columns=["AcceptedCmp1" , "AcceptedCmp2", "AcceptedCmp3" , "AcceptedCmp4","AcceptedCmp5", "Response","NumWebVisitsMonth", "NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumDealsPurchases" , "Kidhome", "Teenhome","MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"],axis=1)
```

```python
dupl_df
```

```python
#AGE
dupl_df["Age"]=2021 - dupl_df["Year_Birth"]
```

```python
dupl_df['Education'].value_counts()
dupl_df['Education'] = dupl_df['Education'].replace(['PhD','2n Cycle','Graduation', 'Master'],'PG')  
dupl_df['Education'] = dupl_df['Education'].replace(['Basic'], 'UG')
```

```python
dupl_df['Years_Completed'] = 2021- dupl_df['Dt_Customer'].dt.year 
```

## Pair plot for numerical columns only

```python

numerical_cols = dupl_df.select_dtypes(include='number').columns
px.scatter_matrix(dupl_df[numerical_cols])
```

## 8.

```python
correlation_matrix = dupl_df[numerical_cols].corr()

# Find top 3 positive correlation pairs
top_3_positive_corr = correlation_matrix.unstack().sort_values(ascending=False)
top_3_positive_corr = top_3_positive_corr[top_3_positive_corr < 1].drop_duplicates().head(3)

# Create scatter plots for the top 3 positive correlation pairs using Plotly
for i in range(0, len(top_3_positive_corr)):
    col1, col2 = top_3_positive_corr.index[i]
    corr_value = top_3_positive_corr[col1, col2]
    fig = px.scatter(dupl_df, x=col1, y=col2, title=f"Scatter plot: {col1} vs {col2} (Correlation: {corr_value:.4f})")
    fig.show()
```

```python
correlation_matrix = dupl_df[numerical_cols].corr()

# Find top 3 negative correlation pairs
top_3_positive_corr = correlation_matrix.unstack().sort_values(ascending=True)
top_3_positive_corr = top_3_positive_corr[top_3_positive_corr < 0].drop_duplicates().head(3)

# Create scatter plots for the top 3 negative correlation pairs using Plotly
for i in range(0, len(top_3_positive_corr)):
    col1, col2 = top_3_positive_corr.index[i]
    corr_value = top_3_positive_corr[col1, col2]
    fig = px.scatter(dupl_df, x=col1, y=col2, title=f"Scatter plot: {col1} vs {col2} (Correlation: {corr_value:.4f})")
    fig.show()
```
